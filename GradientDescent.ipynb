We are given the following multi-variate function:

\begin{equation}
    g(\mathbf{w}) = w_1^2-4w_1 + w_2^2-8w_2+w_3^2
    \label{eq:1}
\end{equation}

Optimize (minimize) $g(\mathbf{w})$ by manually executing the Gradient Descent (GD) algorithm. Use the following parameters:
- Number of iterations: $3$.
- Learning rate: $\alpha=0.1$.
- Starting point: $\mathbf{w}^{(0)}=(1,1,1)$.


## Solution

### Implementation detail

During our calculations, we will use two special arrays, `weight_history` and `cost_history`. The first one will store the co-ordinates of the points $\mathbf{w}^{(0)}, \mathbf{w}^{(1)}, \mathbf{w}^{(2)},\dots$ that are computed by each GD iteration. On the other hand, `cost_history` will store the values of the given function $g\big(\mathbf{w}^{(0)}\big), g\big(\mathbf{w}^{(1)}\big), g\big(\mathbf{w}^{(2)}\big),\dots$ at the aforementioned points.

**Nevertheless, notice that both of these arrays are optional**. In fact, none of them is really necessary to apply the Gradient Descent algorithm. Therefore, they can be simply omitted and the algorithm will work exactly the same. In this example, we create and update them merely for "educational" purposes, so that the reader may realize how the points and the cost function itself get optimized.

Consequently, at the beginning, before the execution of GD we can insert:

1. the starting point $\mathbf{w}^{(0)}=(1,1,1)$ into the `weight_history` array, and
2. the value of the function $g\big(\mathbf{w}^{(0)}\big)$ at the starting point $\mathbf{w}^{(0)}$ into the `cost_history` array. We can calculate it easily by setting the co-ordinates of the starting point (i.e., $w_1=1$, $w_2=1$, $w_3=1$) into Eq. \eqref{eq:1}:

\begin{equation}
g\big(\mathbf{w}^{(0)}\big) = w_1^2-4w_1 + w_2^2-8w_2+w_3^2 = 1^2 - 4\cdot 1 + 1^2 - 8\cdot 1 + 1^2 = -9.0.
\end{equation}


### Preparation step

At first, we observe that the given function $g(\mathbf{w})$ includes 3 variables: $w_1$, $w_2$, $w_3$. Consequently, **each point that lies on the curve of the function $g(\mathbf{w})$ will be three-dimensional**: $(w_1,w_2,w_3)$. We can also verify this by observing the given starting point $\mathbf{w}^{(0)}=(1,1,1)$ which is also three-dimensional.

We start the solution by manually calculating the gradient of the given function, that is $\nabla g(\mathbf{w})$. So:

\begin{equation}
\nabla g(\mathbf{w})=\begin{bmatrix} \frac{\partial g}{\partial w_1} \\ \frac{\partial g}{\partial w_2} \\ \frac{\partial g}{\partial w_3} \end{bmatrix} = \begin{bmatrix} 2w_1 - 4 \\ 2w_2 - 8 \\ 2w_3 \end{bmatrix}
\label{eq:grad}
\end{equation}


Eq. \eqref{eq:grad} provides the gradient vector of $g(\mathbf{w})$. We will now use this form to iteratively optimize the function, according to the GD algorithm.


### Iteration 1

According to the generic form of the GD algorithm, we can "improve" $\mathbf{w}^{(0)}$ by finding another (better) point $\mathbf{w}^{(1)}$ that derives from the following equation:

\begin{equation}
\mathbf{w}^{(1)}=\mathbf{w}^{(0)} - \alpha \nabla g\big(\mathbf{w}^{(0)}\big)
\label{eq:gd1}
\end{equation}

We say that $\mathbf{w}^{(1)}$ is improved (it is better) in relation to $\mathbf{w}^{(0)}$, beacuse after the execution of the first iteration, it will be valid that $g\big(\mathbf{w}^{(1)}\big) < g\big(\mathbf{w}^{(0)}\big)$. Therefore, all we have to do is to plug Eq. \eqref{eq:grad} to Eq. \eqref{eq:gd1} and place the co-ordinates of $\mathbf{w}^{(0)}=(1,1,1)$:

\begin{equation}
\mathbf{w}^{(1)}=\mathbf{w}^{(0)} - \alpha \nabla g(\mathbf{w}^{(0)})=\begin{bmatrix} w_1^{(0)} \\ w_2^{(0)} \\ w_3^{(0)} \end{bmatrix} - \alpha\begin{bmatrix} 2w_1^{(0)} - 4 \\ 2w_2^{(0)} - 8 \\ 2w_3^{(0)} \end{bmatrix}=\begin{bmatrix} 1 \\ 1 \\ 1 \end{bmatrix} - 0.1\begin{bmatrix} 2\cdot 1 - 4 \\ 2\cdot 1 - 8 \\ 2\cdot 1 \end{bmatrix} = \begin{bmatrix} 1+0.2\\ 1+0.6 \\1-0.2 \end{bmatrix}=\begin{bmatrix} 1.2\\ 1.6 \\0.8 \end{bmatrix}
\label{eq:p1}
\end{equation}

Consequently, the new point in the function's curve is $\mathbf{w}^{(1)}=(1.2, 1.6, 0.8)$. We can (optionally) append this point into the `weight_history` array. Now let us compute the value of the function $g(\mathbf{w})$ at the point $\mathbf{w}^{(1)}$:

\begin{equation}
g\big(\mathbf{w}^{(1)}\big) = w_1^2-4w_1 + w_2^2-8w_2+w_3^2 = 1.2^2 - 4\cdot 1.2 + 1.6^2 - 8\cdot 1.6 + 0.8^2 = -12.96.
\end{equation}

We can (optionally) append the value of the function $g\big(\mathbf{w}^{(1)}\big)$ into the `cost_history` array. It is valid that $g\big(\mathbf{w}^{(1)}\big) < g\big(\mathbf{w}^{(0)}\big)$.


### Iteration 2

We can further improve $\mathbf{w}^{(1)}$ as follows:

\begin{equation}
\mathbf{w}^{(2)}=\mathbf{w}^{(1)} - \alpha \nabla g(\mathbf{w}^{(1)})
\label{eq:gd2}
\end{equation}

All we have to do is to plug Eq. \eqref{eq:grad} to Eq. \eqref{eq:gd2} and place the co-ordinates of $\mathbf{w}^{(1)}=(1.2, 1.6, 0.8)$, (See Eq. \eqref{eq:p1}):

\begin{equation}
\mathbf{w}^{(2)}=\mathbf{w}^{(1)} - \alpha \nabla g(\mathbf{w}^{(1)}) = \begin{bmatrix} w_1^{(1)} \\ w_2^{(1)} \\ w_3^{(1)} \end{bmatrix} - \alpha\begin{bmatrix} 2w_1^{(1)} - 4 \\ 2w_2^{(1)} - 8 \\ 2w_3^{(1)} \end{bmatrix}=\begin{bmatrix} 1.2 \\ 1.6 \\ 0.8 \end{bmatrix} - 0.1\begin{bmatrix} 2\cdot 1.2 - 4 \\ 2\cdot 1.6 - 8 \\ 2\cdot 0.8 \end{bmatrix} = \begin{bmatrix} 1.2 +0.16\\ 1.6+0.48 \\0.8-0.16 \end{bmatrix}=\begin{bmatrix} 1.36\\ 2.08 \\0.64 \end{bmatrix}
\label{eq:p2}
\end{equation}

Consequently, our new point is $\mathbf{w}^{(2)}=(1.36, 2.08, 0.64)$ and we can optionally append it to the `weight_history` array. Now let us compute the value of the function $g(\mathbf{w})$ at the point $\mathbf{w}^{(2)}$:

\begin{equation}
g(\mathbf{w}^{(2)}) = w_1^2-4w_1 + w_2^2-8w_2+w_3^2 = 1.36^2 - 4\cdot 1.36 + 2.08^2 - 8\cdot 2.08 + 0.64^2 = -15.4944.
\end{equation}

We can (optionally) append the value of the function $g\big(\mathbf{w}^{(2)}\big)$ into the `cost_history` array. It is valid that $g\big(\mathbf{w}^{(2)}\big) < g\big(\mathbf{w}^{(1)}\big) < g\big(\mathbf{w}^{(0)}\big)$. 


### Iteration 3

We can further improve $\mathbf{w}^{(2)}$ as follows:

\begin{equation}
\mathbf{w}^{(3)}=\mathbf{w}^{(2)} - \alpha \nabla g(\mathbf{w}^{(2)})
\label{eq:gd3}
\end{equation}

All we have to do is to plug Eq. \eqref{eq:grad} to Eq. \eqref{eq:gd3} and place the co-ordinates of $\mathbf{w}^{(2)}=(1.36, 2.08, 0.64)$, (See Eq. \eqref{eq:p2}):

\begin{equation}
\mathbf{w}^{(3)}=\mathbf{w}^{(2)} - \alpha \nabla g(\mathbf{w}^{(2)}) = \begin{bmatrix} w_1^{(2)} \\ w_2^{(2)} \\ w_3^{(2)} \end{bmatrix} - \alpha\begin{bmatrix} 2w_1^{(2)} - 4 \\ 2w_2^{(2)} - 8 \\ 2w_3^{(2)} \end{bmatrix}=\begin{bmatrix} 1.36 \\ 2.08 \\ 0.64 \end{bmatrix} - 0.1\begin{bmatrix} 2\cdot 1.36 - 4 \\ 2\cdot 2.08 - 8 \\ 2\cdot 0.64 \end{bmatrix} = \begin{bmatrix} 1.36+0.128 \\ 2.08+0.384 \\0.64-0.128 \end{bmatrix}=\begin{bmatrix} 1.488\\ 2.464 \\0.512\end{bmatrix}
\label{eq:p3}
\end{equation}

Consequently, our new point is $\mathbf{w}^{(3)}=(1.488, 2.464, 0.512)$ and we can optionally append it to the `weight_history` array. Now let us compute the value of the function $g(\mathbf{w})$ at the point $\mathbf{w}^{(3)}$:

\begin{equation}
g(\mathbf{w}^{(3)}) = w_1^2-4w_1 + w_2^2-8w_2+w_3^2 = 1.488^2 - 4\cdot 1.488 + 2.464^2 - 8\cdot 2.464 + 0.512^2 = -17.116416.
\end{equation}

We can (optionally) append the value of the function $g\big(\mathbf{w}^{(3)}\big)$ into the `cost_history` array. It is valid that $g\big(\mathbf{w}^{(3)}\big) < g\big(\mathbf{w}^{(2)}\big) < g\big(\mathbf{w}^{(1)}\big) < g\big(\mathbf{w}^{(0)}\big)$.


## Implementation (with autograd)

The minimum implementation of Gradient Descent is by using the `autograd` library. Recall that `autograd` is a powerful NumPy extension that enables the automatic calculation of function gradients. So, the simplest implementation of GD with `autograd` is:


```python
# Two imports
import autograd.numpy as np
from autograd import grad

# Gradient Descent with autograd.
# Inputs: g (input function), alpha (steplength parameter), iters (number of iterations), w (initial point)
def gradient_descent(g, alpha, iters, w):
    # compute gradient vector of the input function g by using autograd
    gradient = grad(g)

    # run the gradient descent loop
    weight_history = [w]           # container for weight history
    cost_history = [g(w)]          # container for corresponding cost function history
    
    for k in range(iters):
        # use the gradient vector to evaluate the gradient at the current point w
        grad_at_w = gradient(w)

        # update (i.e. find a better) point 
        w = w - alpha * grad_at_w

        # record weight and cost
        weight_history.append(w)
        cost_history.append(g(w))

    return weight_history,cost_history
```

Now let us use this implementation for our example cost function $g(\mathbf{w})$:

```python
# A lambda expression for g
g = lambda w: w[0]**2 - 4*w[0] + w[1]**2 - 8*w[1] + w[2]**2

# Initial Point
w0 = np.array([1.0, 1.0, 1.0])

# Learning rate
alpha = 0.1;

# Iterations
iterations = 3;

# execute Gradient Descent
weight_history, cost_history = gradient_descent(g, alpha, iterations, w0)

print("Weight History:\n", weight_history)
print("Cost History:\n", cost_history)
```

## Steplength study

The cost function of Eq. \eqref{eq:1} was ideal to perceive how Gradient Descent works and develop its implementation code. However, the three-dimensional functions are not easy (if not impossible) to visualize. For this reason, in this exercise we will apply GD to minimize a simple univariate function. The function is defined as follows:

\begin{equation}
g(\mathbf{w}) = \frac{1}{50}\left(w^4 + w^2 + 10w \right)
\end{equation}

with a starting point $\mathbf{w}^{(0)}=(2)$ and $100$ iterations.


### Fixed steplengths

We will perform three separate runs of the GD code that we developed earlier, by using three different steplength values, namely, $\alpha = 1$, $\alpha = 10^{-1}$, and $\alpha=10^{-2}$. In the sequel, we will plot the `cost_history` array of $g(\mathbf{w})$ for each run in a single figure to compare their performance and identify the steplength value that works better on this particular function.

The function $g(\mathbf{w})$ is a univariate function, so its gradient vector is essentially determined by its first-order derivative. Basic calculus provides this derivative:

\begin{equation}
\frac{\partial}{\partial w}g(w) = \frac{1}{50}\left(4w^3 + 2w + 10 \right)
\label{eq:deriv}
\end{equation}

Nevertheless, the `autograd` library automatically computes the required gradients, therefore Eq. \eqref{eq:deriv} is not actually required.


```python
# A lambda expression for g
g = lambda w: 1 / 50 * (w**4 + w**2 + 10*w)

# Initial Point
w0 = np.array([2.0])

# Iterations
iterations = 100;

# Learning rate 1
alpha = 1;
weight_history_1, cost_history_1 = gradient_descent(g, alpha, iterations, w0)

# Learning rate 2
alpha = 10**(-1);
weight_history_2, cost_history_2 = gradient_descent(g, alpha, iterations, w0)

# Learning rate 3
alpha = 10**(-2);
weight_history_3, cost_history_3 = gradient_descent(g, alpha, iterations, w0)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(18, 6))

# Compute the x and y coordinates for points on sine and cosine curves
x = np.arange(0, iterations + 1, 1.0)

# Plot the points using matplotlib
plt.plot(x, cost_history_1)       # plot cost_history_1
plt.plot(x, cost_history_2)       # plot cost_history_2
plt.plot(x, cost_history_3)       # plot cost_history_3

# Axes formatting
plt.xlabel('Iteration')           # horizontal axis label
plt.ylabel('cost_history')        # vertical axis label
plt.legend([r'$\alpha=1$', r'$\alpha=0.1$', r'$\alpha=0.01$']) # legend

plt.show()
```

### Diminishing steplengths

Setting diminishing steplengths is a common technique in function optimization applications. In this case, we set an initial value for $\alpha$, and at each GD iteration $k$, we decrease it by a small factor that is usually set equal to $\alpha/k$. To support this functionality, we have to implement a very slight change to our `gradient_descent` implementation.

In the following slightly modified version, we introduce an additional parameter `alpha_type` that controls the steplength. If `alpha_type` is set to `diminishing`, then the steplength changes as we discussed earlier. Otherwise, the fixed value of `alpha` is applied, similarly to the previous examples.


```python
# Gradient Descent with autograd and support for diminishing steplengths.
# Inputs: g (input function), alpha_type (diminishing/fixed), alpha (steplength parameter),
# iters (number of iterations), w (initial point)
def gradient_descent(g, alpha, iters, w, alpha_type = 'fixed'):
    # compute gradient vector of the input function g by using autograd
    gradient = grad(g)

    # run the gradient descent loop
    weight_history = [w]           # container for weight history
    cost_history = [g(w)]          # container for corresponding cost function history
    
    for k in range(iters):
        # use the gradient vector to evaluate the gradient at the current point w
        grad_at_w = gradient(w)

        if alpha_type == 'diminishing':
            alpha = alpha  / (k + 1.0)

        # update (i.e. find a better) point 
        w = w - alpha * grad_at_w

        # record weight and cost
        weight_history.append(w)
        cost_history.append(g(w))

    return weight_history,cost_history
```

```python
weight_history_4, cost_history_4 = gradient_descent(g, 1.0, iterations, w0, 'diminishing')

fig = plt.figure(figsize=(18, 6))

# Plot the points using matplotlib
plt.plot(x, cost_history_1)       # plot cost_history_1
plt.plot(x, cost_history_2)       # plot cost_history_2
plt.plot(x, cost_history_3)       # plot cost_history_3
plt.plot(x, cost_history_4)       # plot cost_history_4

# Axes formatting
plt.xlabel('Iteration')           # horizontal axis label
plt.ylabel('cost_history')        # vertical axis label
plt.legend([r'$\alpha=1$', r'$\alpha=0.1$', r'$\alpha=0.01$', r'Diminishing, $\alpha=1/k$']) # legend

plt.show()
```

The Figure above illustrates how the diminishing steplength (red curve) works. The algorithm converges much faster, however, it fails at dinding (i.e. it overshoots) the function's global minimum.


## Gradient oscillations

In this subsection we demonstrate a peculiar case where the gradients of a function oscillate around the minimum without being able to reach it. We are given the following cost function:

\begin{equation}
g(\mathbf{w}) = w_0^2+w_1^2+2\sin (1.5(w_0+w_1)) + 2
\end{equation}

We will apply Gradient Descent with $10$ iterations, a starting point of $\mathbf{w} = (3.0,3.0)$, and three different steplength values: $\alpha = 0.01$, $\alpha = 0.1$, and $\alpha = 1$.


```python
# A lambda expression for g
g = lambda w: w[0]**2 + w[1]**2 + 2 * np.sin(1.5*(w[0] + w[1])) + 2

# Initial Point
w0 = np.array([3.0, 3.0]);

# Iterations
iterations = 10;

# Learning rate 1
alpha = 1;
weight_history_1, cost_history_1 = gradient_descent(g, alpha, iterations, w0)

# Learning rate 2
alpha = 10**(-1);
weight_history_2, cost_history_2 = gradient_descent(g, alpha, iterations, w0)

# Learning rate 3
alpha = 10**(-2);
weight_history_3, cost_history_3 = gradient_descent(g, alpha, iterations, w0)
```

```python
fig = plt.figure(figsize=(18, 6))

# Compute the x and y coordinates for points on sine and cosine curves
x = np.arange(0, iterations + 1, 1.0)

# Plot the points using matplotlib
plt.plot(x, cost_history_1)       # plot cost_history_1
plt.plot(x, cost_history_2)       # plot cost_history_2
plt.plot(x, cost_history_3)       # plot cost_history_3

# Axes formatting
plt.xlabel('Iteration')           # horizontal axis label
plt.ylabel('cost_history')        # vertical axis label
plt.legend([r'$\alpha=1$', r'$\alpha=0.1$', r'$\alpha=0.01$']) # legend

plt.show()
```
