## Variables and data types

The declaration of the data type is implicit and it happens automatically when a value is assigned to a variable.

* Numbers: integer, long integer, float, complex.

## Strings

Strings are sequences of characters, digits, and/or symbols used to store alphanumeric values. In Python, strings adopt zero-based indexing; that is, the first character gets an index equal to 0, the second gets an index equal to 1, etc.

A string slice is a subset of the initial string. More specifically, `c[i:j]` returns a subset of `c` beginning from the `i`-th character and ending (but not including) at the `j`-th character.

**Note:** Slices also apply to lists and tuples.


```python
s = "Machine Learning"
print ("s[0] =", s[0])
print ("s[1] =", s[1])
print ("s[-1] =", s[-1])
print ("s[1:10] =", s[1:10])
print ("s[5:14] =", s[5:14])
```

Python has a rich family of [string manipulation functions](https://docs.python.org/3/library/stdtypes.html#string-methods). A small, but representative subset of these functions includes:

* `st.lower()`, `st.upper()`, `st.casefold()`: These functions convert all characters of `st` to lowercase, uppercase, and lowercase without exceptions, respectively. The latter applies the casefolding algorithm which is a stronger and more aggressive version of `lower()`.
* `len(st)`: returns the length of `st`.
* `st.count(what)`: returns the number of the occurrences of `what` in `st`.
* `st.find(what, pos)`: returns the starting index of `what` within `st`, starting from the `pos`-th character. It returns `-1` if the search is unsuccessful.
* Checking functions: `st.isalnum()`, `st.isalpha()`, `st.isupper()`, `st.isnumeric()`, `st.isdecimal()`, `st.islower()`, etc.
* `st.replace(old, new, count)`: This function replaces all the occurrences of `old` by `new` within the target string `st`. The replacement concerns only the first `count` occurrences of `old`.
* `st.split(sep)`: returns a list that contains all the substrings of `st` delimited by `sep`.
* `st.strip(chars)`: removes all leading and trailing `chars` from `st`.


```python
# String function examples
st = "Python Machine Learning 1 - ελληνικά - der Fluß"
print("original:\t", st) 
print("upper:\t\t", st.upper())      # convert st to uppercase
print("lower:\t\t", st.lower())      # convert st to lowercase
print("casefold:\t", st.casefold())  # perform casefolding to st
print("capitalize:\t", st.capitalize()) # convert the first character of st to upper
print("length:\t\t", len(st))        # returns the length of st
print("count \"a\"'s:\t", st.count("a"))  # how many "a"'s in st?

# Find mutliple occurrences of "a"
firstocc = st.find("a", 0)
secondocc = st.find("a", firstocc + 1)
thirdocc = st.find("a", secondocc + 1)
print(firstocc, secondocc, thirdocc)

# replace "a" by "-a-"
print("replace \"a\"'s':\t", st.replace("a", "-a-"))
```

```python
# String function examples (continued)
st = 'Parallel Programming with Python'
list1 = st.split(' ')
print(list1)
list2 = st.split('P')
print(list2)

st = '     many spaces      '
print("|" + st.strip(' ') + "|")

st = 'www.example.com'
print(st.strip('cmowz.'))
```

## Lists

Lists are data structures that store data elements in an array-like fashion. They are declared by typing comma separated values, enclosed within square brackets. 

The list elements may be of the same, or different types. Therefore, a list may simultaneously contain numbers, strings, objects, or even other lists.

Similarly to strings, the lists also adopt zero-based indexing of their elements (the first element gets an index equal to 0, the second gets an index equal to 1, and so on). In fact, strings are special types of lists that contain characters.

The concatenation and slicing operations are also supported, similarly to strings.


```python
squares = [1, 4, 9, 16, 25]                   # List declaration
newlist = squares + [36, 49, 64, 81, 100]     # List concatenation
print("newlist:\t", newlist)

cubes = [1, 8, 27, 65, 125]                   # Another List
cubes[3] = 64                                 # Direct access by index
cubes.append(216)                             # Append another element
print("cubes:\t\t", cubes)

letters = ['a', 'b', 'c', 'd', 'e', 'f', 'g'] # Another List
letters[2:5] = ['C', 'D', 'E']                # Replace some values directly
print("letters:\t", letters)

listlength = len(letters)                     # Get the length of the list
print("listlength:\t", listlength)

mixed = [squares, letters]                    # a list of lists
print("mixed:\t\t", mixed)
print(mixed[0])                               # get the first list
print(mixed[0][2], mixed[1][3])               # get the third element (index 2) from list 1 & the fourth element from list 2
```

```python
# Iterating through lists
squares = [1, 4, 9, 16, 25]                   # List declaration
print("Item count in the list:", len(squares))
i = 0
for square in squares:
    print("Element Index:", i)
    print("Element Value:", square, "or equivalently:", squares[i])
    i += 1
```

In addition, Python has a special set of [list manipulation functions](https://docs.python.org/3/tutorial/datastructures.html). A small representative sample of them includes:

* `lst.reverse()`: it reverses the elements of `lst`.
* `len(lst)`: it returns the length (i.e., number of elements) of `lst`.
* `lst.insert(pos, s)`: it inserts `s` in the `pos`-th index of `lst`.
* `lst.sort(key, reverse=True/False)`: it sorts the elements of `lst`.
* `lst.append(s)`: it inserts `s` in the end of `lst`.
* `lst.remove(s)`: it removes `s` from `lst`.
* `lst.clear()`: It removes all elements from `lst`.


```python
# list useful methods
squares = [1, 4, 9, 16, 25]                   # List declaration

squares.reverse()                             # Reverse the items of squares
print("Reversed:\t", squares)

squares.insert(2, 36)                         # Insert 36 at the position with index 2 in squares
print("Inserted 36:\t", squares)

squares.sort(reverse=False)                   # Sort squares
print("Sorted:\t\t", squares)

squares.append(49)                            # Insert 49 at the end of squares
print("Appended 49:\t", squares)

squares.remove(9)                             # Remove 9 from the list
print("Removed 9:\t", squares)

squares.clear()                               # Remove all elements from the list
print("Cleared:\t", squares)
```

```python
# Sort by a given key
def takeSecond(elem):
    return elem[1]

# random list
random = [(2, 2), (3, 4), (4, 1), (1, 3)]

# sort list with key
random.sort(key=takeSecond)


# print list
print('Sorted list:', random)
```

## Dictionaries

Python Dictionaries are unordered collections of items. Each item in a dictionary is represented by a key/value pair. Dictionaries are optimized to retrieve values when the key is known. They are declared by typing comma separated values, enclosed within curly braces `{}`.

A Dictionary item consists of a key and a corresponding value; it is expressed in the form of a `(key: value)` pair. **While the values can be data of any type and can be duplicated, keys must be of immutable type (i.e. string, number, or tuple with immutable elements) and must be unique**.

In contrast to lists and tuples that use indexes to access their values, a dictionary uses the keys of its elements. Keys can be accessed either by using a square bracket syntax `[]`, or by invoking the `get()` method.


```python
# An empty dictionary
d = {}
print(d)

# A dictionary with integer keys
d = {1: 'Monday', 2: 'Tuesday', 3: 'Wednesday', 4: 'Thursday', 5: 'Friday'}
print(d)

# A dictionary with mixed keys (a string and an int)
d = {'day': 'Monday', 1: [2, 4, 3]}
print(d)

# using dict()
d = dict( {1: 'Monday', 2: 'Tuesday', 3: 'Wednesday', 4: 'Thursday', 5: 'Friday'} )
print(d)

# from sequence having each item as a pair
d = dict( [ (1, 'Monday'), (2, 'Tuesday'), (3, 'Wednesday'), (4, 'Thursday'), (5, 'Friday') ] )
print(d)
```

```python
# get vs [] for retrieving elements
d = {'name': 'Jack', 'age': 26}

# Output: Jack
print(d['name'])

# Output: 26
print(d.get('age'))

# Trying to access keys which doesn't exist throws error
# print(my_dict.get('address'))

# KeyError
#print(my_dict['address'])
```

The Dictionaries are mutable. We can add new items into the data structure, or change the value of existing items using an assignment operator.

If the key is already present, then the corresponding value gets updated. In case the key is not present, a new `(key: value)` pair is inserted into the dictionary.


```python
# Changing and adding Dictionary Elements
d = {'name': 'Jack', 'age': 26}
print("Original Dictionary:", d)

# update value
d['age'] = 27

#Output: {'age': 27, 'name': 'Jack'}
print("Updated Dictionary:", d)

# add item
d['address'] = 'Downtown'
print("Updated Dictionary:", d)

# Output: {'address': 'Downtown', 'age': 27, 'name': 'Jack'}

```

## Flow control

The flow of a Python program can be controlled through standard `if-elif-else` statements.

In contrast to the typical C-like syntax, the `if` statements include no curly braces. In Python, the various blocks of code are determined by the indentation (namely, the left whitespace) of the block.

Unlike the C-like syntax, the simple `if` conditions are not parenthesized. `if` conditions may support more complex expressions, which may require parentheses.


```python
x = int(input("Please enter an integer: "))

if x < 0:
    x = 0
    print('Negative changed to zero')
elif x == 0:
    print('Zero')
else:
    print('Positive x')
```

```python
x = int(input("Please enter an integer: "))
y = int(input("Please enter another integer: "))
z1 = 20
z2 = -10
if x < 0 and y < 0:
    x = -x
    y = -y
    print('Both x and y are negative and have been changed to positive')
elif x == 0 or y == 0:
    print('At least one of x OR y are zero')
elif x > 0 and y > 0:
    print('Both x and y are positive')
elif (x < 0 and y < 0) or (z1 > 0 and z2 > 0):
    print('Executing the complex statement')
else:
    print('Nothing special')
```

## Handling files

Python can read from and write to standard text and binary files. Two pieces of information are required:

* a valid filename that corresponds to an existing location in the hard drive (with the appropriate read/write perimissions).
* the file access mode that determines the purpose of opening a file:
    * `r`: Read - Default value. Opens a file for reading. An error is triggered if the file does not exist.
    * `a`: Append - Opens a file for appending data. If the file does not exist, it will be automatically created.
    * `w`: Write - Opens a file for writing data. If the file does not exist, it will be automatically created.
    * `x`: Create - Creates the specified file. An error is triggered if another file with the same name already exists.


```python
# Writing data to a file
print("Writing data to a file")
f = open('test.txt', 'w')
f.write("Line 1: writing to a test file!\n")
f.write("Line 2: writing a second line to a test file!\n")
f.close()

# Reading data from a file
print("Reading data from a file")
f = open('test.txt', 'r')
t = f.read()
print("Data read:", t)
f.close()

# Reading data from a file line-by-line into a list
print("Reading the file line-by-line")
lines = []
with open('test.txt') as file:
    for line in file:
        lines.append(line)
print(lines)
```

## Classes

Python is an object oriented programming language. Almost everything in Python is an object, with its own properties and methods. [Classes](https://docs.python.org/3/tutorial/classes.html) provide a means of bundling data and functionality together. They consist of attributes and methods that are called member attributes and member methods, respectively. Classes are declared by using the reserved word `class`. When a class definition is entered, a new namespace is created, and used as the local scope - thus, all assignments to local variables go into this new namespace. In particular, function definitions bind the name of the new function here.

The class constructor is a special method that automatically initiates an object upon creation. `__init__` is a special function name that denotes a class constructor. A class constructor may receive an arbitrary number of arguments. Python does not allow constructor overloading.

A class may include multiple member attributes and methods. In the following example, `x` is an object of type `Complex`.
`x` is immediately initialized upon its creation, via the constructor. Its methods are accessible through the dot (`.`)operator.


```python
import math

class Complex:
    # Class constructor - it has two arguments (the 'self' keyword does not count)
    def __init__(self, realpart, imagpart):
        # r and i are member attributes and can be accessed with self.r and self.i respectively
        self.r = realpart  
        self.i = imagpart

    # Member method
    def amp(self):
        return math.sqrt(self.r ** 2 + self.i ** 2)

# x is an object of Complex type.
# We create it by passing 2 arguments, in accordance to the constructor decleration.
x = Complex(3, 4)

print(x.amp())
```

```python
class Student(object):

    institution = "IHU"        # class variable shared by all instances (similar to static C++/Java class members)

    def __init__(self, name, dep):
        self.name = name       # instance variable unique to each instance
        self.department = dep  # instance variable unique to each instance

stu1 = Student("Bob", "Data science")
stu2 = Student("Alice", "Mobile computing")

print(stu1.name, ":", stu1.institution)   # shared by all Students
print(stu2.name, ":", stu2.institution)   # shared by all Students

print(stu1.name, ":", stu1.department)    # unique to Bob
print(stu2.name, ":", stu2.department)    # unique to Alice
```

### Inheritance

Python classes provide all the standard features of Object Oriented Programming. Therefore, the class inheritance mechanism:
* allows multiple base classes,
* a derived class can override any methods of its base class or classes, and
* a method can call the method of a base class with the same name.

Objects can contain arbitrary amounts and kinds of data. As is true for modules, classes partake of the dynamic nature of Python: they are created at runtime, and can be modified further after creation.


```python
class Person:
    def __init__(self, fname, lname):
        self.firstname = fname
        self.lastname = lname

    def printname(self):
        print(self.firstname, self.lastname)

# Use the Person class to create an object,
# and then execute the printname method:
x = Person("John", "Doe")
x.printname()

class Employee(Person):
    def __init__(self, fname, lname, p, s):
        Person.__init__(self, fname, lname)
        self.salary = s
        self.position = p

    def printall(self):
        print("Employee Data",
              "\nFirst:", self.firstname,
              "\nLast:", self.lastname,
              "\nPosition:", self.position,
              "\nSalary:", self.salary)

y = Employee("Michael", "Owen", "Project Manager", 2000)
y.printall()
```
